```python
import pandas as pd
import re
```

```python
data=pd.read_csv('link_removal_dataset.csv')
```

```python
data.head()
```

```python
data['text']
```

```python
def remove_links(text):
    pattern=re.compile(r'https?://\S+|WWW\S+')
    return pattern.sub(r'',text)
```

```python
text="This tutorial is available at http://www.learnnlp.com/course'"
remove_links(text)
```

```python
data['text'].apply(remove_links)
```
