```python
import string ,time
string.punctuation
```

```python
exclude=string.punctuation
```

```python
def remove_punc(text):
    for char in exclude:
        text=text.replace(char,'')
    return text
```

```python
text='string.with.punction?'
```

```python
print(remove_punc(text))
```

```python
start=time.time()
print(remove_punc(text)) #fir this method takes time 
time1=time.time()-start
print(time1*50000)
```

```python
def remove(text):
    return text.translate(str.maketrans('','',exclude))
```

```python
start=time.time()
remove(text)
time2=time.time()-start #require less time 
print(time2*50000)
```
