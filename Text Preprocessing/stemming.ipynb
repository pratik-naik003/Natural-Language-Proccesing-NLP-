```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample='walk walks walking walked'
```

```python
stem_words(sample)
```

```python
import pandas as pd
data=pd.read_csv('stemming_dataset_nltk.csv')

data.head()
```
