```python
#function to remove stopwords
def remove_words(text):
    new_text=[]
    
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x=new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
import pandas as pd 
```

```python
data=pd.read_csv('stopwords_removal_dataset.csv')
data
```

```python
data['original_text']=data['original_text'].apply(remove_words)
data
```
