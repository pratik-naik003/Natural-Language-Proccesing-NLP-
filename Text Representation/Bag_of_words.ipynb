```python
import pandas as pd

data=pd.read_csv('bag_of_words_dataset.csv')
data.head()
```

```python
data['text']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
x=data['text']
vec=CountVectorizer()
y=vec.fit_transform(x)
print(y.toarray())
```

#some important parametrs

```python
#lowercase 
from sklearn.feature_extraction.text import CountVectorizer
x=data['text']
vec=CountVectorizer(lowercase=True)
y=vec.fit_transform(x)
print(y.toarray())
```

Handling stop words

```python
from sklearn.feature_extraction.text import CountVectorizer
x=data['text']
vec=CountVectorizer(stop_words='english')
y=vec.fit_transform(x)
print(y.toarray())
```

Max features

```python
from sklearn.feature_extraction.text import CountVectorizer
x=data['text']
vec=CountVectorizer(max_features=6)
y=vec.fit_transform(x)
print(y.toarray())
```
