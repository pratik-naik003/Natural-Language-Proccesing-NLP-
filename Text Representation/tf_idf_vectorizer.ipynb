```python
import pandas as pd

data=pd.read_csv('tfidf_dataset.csv')
data
```

```python
data['text']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

x=data['text']
vec=TfidfVectorizer()
y=vec.fit_transform(x)
y.toarray()
```
